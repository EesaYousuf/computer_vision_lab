```python
import matplotlib.pyplot as plt

# Group the data into a list of semesters
data = [sem1, sem2, sem3, sem4, sem5, sem6, sem7, sem8]

# Create the box plot
plt.boxplot(data)

# Add labels for each semester
plt.xticks(ticks=range(1, 9), labels=[f"Sem {i}" for i in range(1, 9)])

# Add title and axis labels
plt.title("Box Plot of Subject Marks Across Semesters")
plt.xlabel("Semester")
plt.ylabel("Marks")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data setup
sem1 = [78, 85, 90, 87, 76]
sem2 = [82, 80, 85, 88, 79]
sem3 = [90, 92, 88, 84, 86]
sem4 = [75, 70, 80, 72, 78]
sem5 = [88, 90, 91, 85, 89]
sem6 = [70, 72, 74, 76, 78]
sem7 = [95, 92, 94, 91, 93]
sem8 = [60, 65, 63, 67, 62]

data = {
    'Semester': ['Sem 1']*5 + ['Sem 2']*5 + ['Sem 3']*5 + ['Sem 4']*5 +
                ['Sem 5']*5 + ['Sem 6']*5 + ['Sem 7']*5 + ['Sem 8']*5,
    'Marks': sem1 + sem2 + sem3 + sem4 + sem5 + sem6 + sem7 + sem8
}

df = pd.DataFrame(data)

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Semester', y='Marks', data=df)
plt.title("Violin Plot of Marks Across Semesters")
plt.show()
```

```python
# Swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Semester', y='Marks', data=df)
plt.title("Swarm Plot of Marks Across Semesters")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for 8 semesters
sem1 = [78, 85, 90, 87, 76]
sem2 = [82, 80, 85, 88, 79]
sem3 = [90, 92, 88, 84, 86]
sem4 = [75, 70, 80, 72, 78]
sem5 = [88, 90, 91, 85, 89]
sem6 = [70, 72, 74, 76, 78]
sem7 = [95, 92, 94, 91, 93]
sem8 = [60, 65, 63, 67, 62]

# Combine into long-form DataFrame
data = {
    'Semester': ['Sem 1']*5 + ['Sem 2']*5 + ['Sem 3']*5 + ['Sem 4']*5 +
                ['Sem 5']*5 + ['Sem 6']*5 + ['Sem 7']*5 + ['Sem 8']*5,
    'Marks': sem1 + sem2 + sem3 + sem4 + sem5 + sem6 + sem7 + sem8
}

df = pd.DataFrame(data)

# Create Violin + Swarm plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Semester', y='Marks', data=df, inner=None, color='lightgray')
sns.swarmplot(x='Semester', y='Marks', data=df, color='black', size=5)

plt.title("Violin + Swarm Plot of Subject Marks Across Semesters")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data: marks in 3 subjects over 8 semesters
df = pd.DataFrame({
    'Math':    [80, 85, 78, 92, 88, 90, 85, 87],
    'Physics': [75, 89, 84, 91, 85, 87, 90, 93],
    'Chem':    [70, 88, 82, 89, 84, 86, 89, 91],
    'Semester': ['Sem1', 'Sem2', 'Sem3', 'Sem4', 'Sem5', 'Sem6', 'Sem7', 'Sem8']
})

# Pair plot
sns.pairplot(df.drop(columns='Semester'))
plt.suptitle("Pair Plot of Subject Marks", y=1.02)
plt.show()
```
