## Drawing the Histogram of this Image

What is an Image Histogram?

 X-axis  Pixel intensity values (0 to 255).

 Y-axis  Number of pixels with that intensity.

 Grayscale Histogram  Shows intensity distribution of a single-channel image.

 Color Histogram  Shows separate distributions for Red, Green, and Blue channels

 For this task, we will use a calcHist([img], [0], None, [256], [0, 256]) function, in which
  - the first parameter is the name of image variable.
  - the number of channels.
  - the mask If set to None, the histogram is computed for the entire image. Otherwise, a mask can specify a region of interest (ROI).
  - histsize specifies the number of bins or histogram size.
  - range specifies the pixel values to consider.

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute histogram
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])

# Plot histogram
plt.figure(figsize=(8, 5))
plt.plot(histogram, color='black')  # Black for grayscale
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.xlim([0, 256])  # Intensity range from 0 to 255
plt.show()
```

```python
# Define color channels for visualization
colors = ('blue', 'green', 'red')
channel_labels = ('B', 'G', 'R')

plt.figure(figsize=(8, 5))

# Compute and plot histograms for each channel
for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])  # Compute histogram
    plt.plot(hist, color=color, label=f"{channel_labels[i]} Channel")  # Plot histogram

# Formatting
plt.title("RGB Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.xlim([0, 256])
plt.legend()
plt.show()
```
