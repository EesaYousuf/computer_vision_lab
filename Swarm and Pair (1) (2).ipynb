```python
# Load the Excel file
file_path = '/content/Dataset .xlsx'  # <-- upload the file to Colab first
xls = pd.ExcelFile(file_path)

# Load the "Soft Computing" sheet
df = pd.read_excel(file_path, sheet_name="Soft Computing")

# Drop irrelevant column and clean data
df_clean = df.drop(columns=['Unnamed: 6'])
df_clean.dropna(subset=['Marks', 'Gender', 'Graduation', 'Residence', 'Age'], inplace=True)

# Convert strings to lowercase and remove spaces
df_clean['Gender'] = df_clean['Gender'].str.strip().str.lower()
df_clean['Graduation'] = df_clean['Graduation'].str.strip().str.lower()
df_clean['Residence'] = df_clean['Residence'].str.strip().str.lower()

# Display first few rows
df_clean.head()
```

```python
# Step 1: Import libraries
import pandas as pd

# Step 2: Load the dataset and select the sheet
file_path = "/content/Dataset .xlsx"  # <-- adjust path if needed
df = pd.read_excel(file_path, sheet_name="Soft Computing")

# Step 3: Drop unnamed/empty columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Step 4: Display original dataset
print("🔹 Original Dataset:")
print(df.head())

# Step 5: Drop rows where Age < 24 (but keep 'Age' column)
df_dropped = df[df["Age"] >= 25]

print("\n🔹 After Dropping Rows with Age < 25:")
print(df_dropped)

# Step 6: Fill missing values
# - Age with median
# - Marks with mean
df_filled = df_dropped.copy()
df_filled["Age"] = df_filled["Age"].fillna(df_filled["Age"].median())
df_filled["Marks"] = df_filled["Marks"].fillna(df_filled["Marks"].mean())

print("\n🔹 After Filling Missing Values:")
print(df_filled)
```

```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = '/content/Dataset .xlsx'  # Adjust this if using locally
df = pd.read_excel(file_path, sheet_name="Soft Computing")


# 💡 Step 1: Replace Age < 25 with NaN (simulate missing data)
df_demo = df.copy()
df_demo.loc[df_demo['Age'] < 25, 'Age'] = np.nan

print("🔹 Age < 24 replaced with NaN:")
print(df_demo[['Name', 'Age']])
```

```python
df_median = df_demo.copy()
df_median['Age'] = df_median['Age'].fillna(df_median['Age'].median())

print("\n🔹 Filled with Median Age:")
print(df_median[['Name', 'Age']])
```

```python
df_mean = df_demo.copy()
df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())

print("\n🔹 Filled with Mean Age:")
print(df_mean[['Name', 'Age']])
```

```python
df_mode = df_demo.copy()
df_mode['Age'] = df_mode['Age'].fillna(df_mode['Age'].mode()[0])

print("\n🔹 Filled with Mode Age:")
print(df_mode[['Name', 'Age']])
```

```python
df_fixed = df_demo.copy()
df_fixed['Age'] = df_fixed['Age'].fillna(30)

print("\n🔹 Filled with Fixed Value (30):")
print(df_fixed[['Name', 'Age']])
```

```python
df_ffill = df_demo.copy()
df_ffill['Age'] = df_ffill['Age'].fillna(method='ffill')

print("\n🔹 Filled using Forward Fill:")
print(df_ffill[['Name', 'Age']])
```

```python
df_bfill = df_demo.copy()
df_bfill['Age'] = df_bfill['Age'].fillna(method='bfill')

print("\n🔹 Filled using Backward Fill:")
print(df_bfill[['Name', 'Age']])
```
