```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Create a 5x5 grayscale matrix (single-channel)
gray_matrix = np.array([
    [0, 50, 100, 150, 200],
    [25, 75, 125, 175, 225],
    [50, 100, 150, 200, 250],
    [75, 125, 175, 225, 255],
    [100, 150, 200, 250, 255]
], dtype=np.uint8)  # Values range from 0 (black) to 255 (white)

# Display as a grayscale image
plt.imshow(gray_matrix, cmap='gray')
plt.title("Grayscale Image from Matrix")
plt.colorbar()
plt.show()
```

```python
# Create 5x5 matrices for each color channel
red_channel = np.array([
    [10, 0, 0, 255, 0],
    [0, 255, 0, 0, 255],
    [255, 0, 255, 0, 0],
    [0, 0, 255, 255, 0],
    [0, 255, 0, 0, 255]
], dtype=np.uint8)  # Red intensity values

green_channel = np.array([
    [0, 255, 0, 0, 255],
    [255, 0, 255, 255, 0],
    [0, 255, 0, 0, 255],
    [255, 255, 0, 0, 255],
    [255, 0, 255, 255, 0]
], dtype=np.uint8)  # Green intensity values

blue_channel = np.array([
    [0, 0, 255, 0, 255],
    [255, 0, 255, 0, 255],
    [0, 255, 255, 255, 0],
    [255, 0, 255, 0, 255],
    [255, 255, 0, 255, 0]
], dtype=np.uint8)  # Blue intensity values

# Stack channels to create an RGB image
color_image = np.stack((red_channel, green_channel, blue_channel), axis=2)

# Display the final RGB image
plt.imshow(color_image)
plt.title("RGB Color Image from Three Matrices")
plt.show()
```
